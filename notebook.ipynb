## Introducci√≥n

La empresa OilyGiant buscar crear un modelo que le ayude a predecir el volumen de petroleo que un pozo puede producir de manera que pueda elegir 200 lugares para extraer el mismo de tres diferentes regiones. A su vez busca estimar los beneficios que podr√≠a esperar a partir de una monto de inversi√≥n que tiene para desarrollar este proyecto.

```python
### Descarga y prepara los datos
```

```python
# Importa las librer√≠as
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from scipy import stats as st
```

```python
# Descarga el Dataset
data_geo_0 = pd.read_csv('/datasets/geo_data_0.csv')
data_geo_1 = pd.read_csv('/datasets/geo_data_1.csv')
data_geo_2 = pd.read_csv('/datasets/geo_data_2.csv')
```

```python
# Imprime la informaci√≥n del Dataset
print(data_geo_0.info())
print(data_geo_1.info())
print(data_geo_2.info())
```

```python
# Imprime 5 primeras filas
print(data_geo_0.head())
print(data_geo_1.head())
print(data_geo_2.head())
```

```python
# Elimina la columna id
data_geo_0 = data_geo_0.drop(['id'], axis=1)
data_geo_1 = data_geo_1.drop(['id'], axis=1)
data_geo_2 = data_geo_2.drop(['id'], axis=1)

print(data_geo_0.head())
print(data_geo_1.head())
print(data_geo_2.head())
```

```python
# Almacena en una variable el nombre de las columnas n√∫mericas
numeric = ['f0', 'f1', 'f2']
#categorical = []
```

```python
# Visualiza las variables a escalar
data_geo_0[numeric].plot(kind="box")
```

```python
# Almacena el escalador en una variable
scaler = StandardScaler()
# Estandariza las variables num√©ricas
data_geo_0[numeric] = scaler.fit_transform(data_geo_0[numeric])
```

```python
# Visualiza las variables escaladas
data_geo_0[numeric].plot(kind="box")
```

### Entrena y prueba el modelo

```python
# Divide las caracter√≠sticas y el objetivo
features = data_geo_0.drop('product',axis=1)
target = data_geo_0['product']

# Divide los datos para entrenamiento y validaci√≥n
features_train, features_valid, target_train, target_valid = train_test_split(features, target, test_size=0.25, random_state=12345)

# Imprime el tama√±o de los datos
print(features_train.shape)
print(features_valid.shape)
```

```python
# inicializa el constructor de modelos
model = LinearRegression()
# entrena el modelo en el conjunto de entrenamiento
model.fit(features_train, target_train)
# obt√©n las predicciones del modelo en el conjunto de validaci√≥n
predictions_valid = model.predict(features_valid)
```

```python
# Obtiene la media de las predicciones
print(predictions_valid.mean(), 'Vol. medio de reservas predicho(miles de barriles).')
# Obtiene el error cuadratico medio
mse = mean_squared_error(target_valid, predictions_valid)
# Obtiene la ra√≠z del error cuadratico medio
rmse = mse ** 0.5

print(rmse, 'RMSE del modelo.')
# Obtiene el error absoluto medio
mae = mean_absolute_error(target_valid, predictions_valid)
print(mae, 'MAE del modelo.')
```

```python
def preprocess (df, mod_target, var_numeric):
    """
    Variables:
        df (Dataset): Dataset to preprocess
        mod_target (string): variable target
        var_numeric (list): List of numerical column names to preprocess

    """
    # Almacena el escalador en una variable
    scaler = StandardScaler()
    # Estandariza las variables num√©ricas
    df[var_numeric] = scaler.fit_transform(df[var_numeric])

    # Divide las caracter√≠sticas y el objetivo
    features = df.drop(mod_target, axis=1)
    target = df[mod_target]

    # Divide los datos para entrenamiento y validaci√≥n
    features_train, features_valid, target_train, target_valid = train_test_split(features, target, test_size=0.25, random_state=12345)
    # inicializa el constructor de modelos
    model = LinearRegression()
    # entrena el modelo en el conjunto de entrenamiento
    model.fit(features_train, target_train)
    # obt√©n las predicciones del modelo en el conjunto de validaci√≥n
    predictions_valid = model.predict(features_valid)

    # Obtiene la media de las predicciones
    print(predictions_valid.mean(), 'Vol. medio de reservas predicho(miles de barriles).')
    # Obtiene el error cuadratico medio
    ecm_mse = mean_squared_error(target_valid, predictions_valid)
    # Obtiene la ra√≠z del error cuadratico medio
    rmse = ecm_mse ** 0.5

    print(rmse, 'RMSE del modelo.')

    # Obtiene el error absoluto medio
    mae = mean_absolute_error(target_valid, predictions_valid)
    print(mae, 'MAE del modelo.')
    # Convierte las predicciones en un Serie
    sample = pd.Series(predictions_valid)
    return sample
```

```python
# Prueba la funci√≥n
sample0 = preprocess(data_geo_0, 'product', numeric)
# Obtiene la estadistica de las predicciones
print(sample0.describe())
# Visualiza la distribuci√≥n de las predicciones
print(sample0.plot(kind='box'))
```

```python
# Prueba la funci√≥n
sample1 = preprocess(data_geo_1, 'product', numeric)
# Obtiene la estadistica de las predicciones
print(sample1.describe())
# Visualiza la distribuci√≥n de las predicciones
print(sample1.plot(kind='box'))
```

```python
# Prueba la funci√≥n
sample2 = preprocess(data_geo_2, 'product', numeric)
# Obtiene la estadistica de las predicciones
print(sample2.describe())
# Visualiza la distribuci√≥n de las predicciones
print(sample2.plot(kind='box'))
```

### Calcula riesgos y ganancias para cada regi√≥n

```python
# Obtiene los 200 mejores pozos
higher_wells_0 = profitable_wells_0.sort_values(ascending=False)[:200]
higher_wells_1 = profitable_wells_1.sort_values(ascending=False)[:200]
higher_wells_2 = profitable_wells_2.sort_values(ascending=False)[:200]

# Parametros de rentabilidad
invesment = 100000000
wells = 200
invesment_per_well = invesment / wells
income_per_unit = 4500
tax_rate = .3

# Covierte el serie en Dataframe
profits_0 = higher_wells_0.to_frame(name='volume')
# Calcula los beneficios y margenes
profits_0['gross_profit'] = profits_0 * income_per_unit
profits_0['operation_profit'] = profits_0['gross_profit'] - invesment_per_well
profits_0['operation_margin'] = profits_0['operation_profit'] / profits_0['gross_profit']
profits_0['net_profit'] = profits_0['operation_profit'] * (1-tax_rate)
profits_0['net_margin'] = profits_0['net_profit'] / profits_0['gross_profit']

print(profits_0)
```

```python
# Declara la funci√≥n
def profit_chart(df):
    # Obtiene los 200 mejores pozos
    higher_wells = df.sort_values(ascending=False)[:200]

    # Parametros de rentabilidad
    invesment = 100000000
    wells = 200
    invesment_per_well = invesment / wells
    income_per_unit = 4500
    tax_rate = .3

    # Covierte el serie en Dataframe
    profits = higher_wells.to_frame(name='volume')
    # Calcula los beneficios y margenes
    profits['gross_profit'] = profits * income_per_unit
    profits['operation_profit'] = profits['gross_profit'] - invesment_per_well
    profits['operation_margin'] = profits['operation_profit'] / profits['gross_profit']
    profits['net_profit'] = profits['operation_profit'] * (1-tax_rate)
    profits['net_margin'] = profits['net_profit'] / profits['gross_profit']
    return profits
```

```python
# Calcula los beneficios y margenes
profits_1 = profit_chart(profitable_wells_1)
print(profits_1)
```

```python
# Calcula los beneficios y margenes
profits_2 = profit_chart(profitable_wells_2)
print(profits_2)
```

```python
# Resumen de beneficios netos por regi√≥n
print('Beneficios netos esperados regi√≥n 0:', profits_0['net_profit'].sum())
print('Beneficios netos esperados regi√≥n 1:', profits_1['net_profit'].sum())
print('Beneficios netos esperados regi√≥n 2:', profits_2['net_profit'].sum())
```
